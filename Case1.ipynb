```python
#Задание:
#Написать программу, которая принимает на вход файл с данными о продажах за определенный период (например, месяц)
#выводит на экран статистику по каждому товару: среднюю цену, общее количество проданных единиц и общую сумму выручки.
#Кроме того, программа должна строить график зависимости выручки от времени для каждого товара.
```

```python
#Подключение библиотек

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as mb
```

```python
b = pd.read_csv('Baza1.csv', delimiter=',')
```

```python
#Выборка по условию(цена)
aa = b.loc[b["product"] == 'apple', "price"]
bb = b.loc[b['product'] == 'banana', "price"]
cc = b.loc[b['product'] == 'cherry', 'price']

#Выборка по условию(кол-во единиц продаж)
aaq = b.loc[b['product'] == 'apple', 'quantity']
bbq = b.loc[b['product'] == 'banana','quantity']
ccq = b.loc[b['product'] == 'cherry', 'quantity']

#Цена за штуку
aa1 = float(np.mean(aa))
bb1 = float(np.mean(bb))
cc1 = float(np.mean(cc))

#Сумма продаж
aa2 = np.sum(aaq)
bb2 = np.sum(bbq)
cc2 = np.sum(ccq)

#Общая сумма выручки продукта
aa3 = np.sum(aa * aaq)
bb3 = np.sum(bb * bbq)
cc3 = np.sum(cc * ccq)

#Создания фрейма с новыми данными(для удобства)
Fruit = pd.DataFrame({
    "Продукт":["Apple", "Banana", "Cherry"],
            "Единица":[aa1, bb1, cc1],
    "Сумма продаж":[ aa2, bb2, cc2],
    "Сумма выручки": [aa3,bb3,cc3]
})

#Вывести фрейм
print(Fruit)
```

```python
# Подготовка данных для создания графика

#Создание списков
one = []
two = []
three = []
dq = [one, two, three]
fruits = ['Apple', 'Banana', 'Cherry']
daate = ['2021-04-01', '2021-04-02', '2021-04-03']


#Выборка выручки по дате и фрукту
aaa1 = b.loc[((b["date"] == '2021-04-01') & (b['product'] == 'apple')), "quantity"]
bbb1 = b.loc[((b["date"] == '2021-04-01') & (b['product'] == 'banana')), "quantity"]
ccc1 = b.loc[((b["date"] == '2021-04-01') & (b['product'] == 'cherry')), "quantity"]

#Выборка выручки по дате и фрукту
aaa2 = b.loc[((b["date"] == '2021-04-02') & (b['product'] == 'apple')), "quantity"]
bbb2 = b.loc[((b["date"] == '2021-04-02') & (b['product'] == 'banana')), "quantity"]
ccc2 = b.loc[((b["date"] == '2021-04-02') & (b['product'] == 'cherry')), "quantity"]

#Выборка выручки по дате и фрукту
aaa3 = b.loc[((b["date"] == '2021-04-03') & (b['product'] == 'apple')), "quantity"]
bbb3 = b.loc[((b["date"] == '2021-04-03') & (b['product'] == 'banana')), "quantity"]
ccc3 = b.loc[((b["date"] == '2021-04-03') & (b['product'] == 'cherry')), "quantity"]

#Добавление выборки в список за '2021-04-01'
one.append(int(aaa1))
one.append(int(bbb1))
one.append(int(ccc1))

#Добавление выборки в список за '2021-04-02'
two.append(int(aaa2))
two.append(int(bbb2))
two.append(int(ccc2))

#Добавление выборки в список за '2021-04-03'
three.append(int(aaa3))
three.append(int(bbb3))
three.append(int(ccc3))



```

```python
# Создание столбчатой диаграммы

# Создание трёх столбцов(основные элементы)
mb.bar(daate, dq[0], color="indianred", align="edge", width=0.2)
mb.bar(daate, dq[1], color="brown", align="edge", width=-0.2)
mb.bar(daate, dq[2], color="darkred", align="center", width=0.2)

# Дополнительные элементы диаграммы
mb.legend(fruits)
mb.xlabel("Дата выручки/продаж")
mb.ylabel("Выручка")
mb.title("Зависимость выручки от времени")
mb.show()







```
